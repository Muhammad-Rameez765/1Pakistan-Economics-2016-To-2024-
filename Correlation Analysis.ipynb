```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv(r"C:\Users\geeze\Downloads\Combined_Data_For_Correlation.csv")
print(df.head())
```

```python

# Time series plot
df.plot(x="Year", y=["Wheat", "Cement", "TotalReserves"], title="Trends Over Time")
plt.show()
```

```python
# Correlation
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
